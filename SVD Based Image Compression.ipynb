## **Code Used**

````python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import numpy as np
from itertools import chain

im1 = Image.open(r"SampleImage5.jpg") 
 
# greyscaling to the original image 
im2 = ImageOps.grayscale(im1) 

title="Original Image"
plt.title(title)
plt.axis('off')
plt.imshow(im2,cmap="gray")
plt.show()


I = np.asmatrix(im2)


[m,n]=I.shape

[U,sigma,V]=np.linalg.svd(I)
r=sigma.size


K=chain(range(2,21,2),range(25,151,25))
K=list(K)

error=[]


for k in K:


    U_k=np.matrix(U[:,0:k])
    sigma_k=np.diag(sigma[0:k])
    V_k=np.matrix(V[0:k,:])
    A=U_k*sigma_k*V_k
    error.append(np.sum(np.square(I-A)))
    
    
    data = Image.fromarray(A)
    title="Singular Values:"+str(k)
    plt.title(title)
    plt.axis('off')
    plt.imshow(data)
    plt.show()
````

````python
plt.plot(K, error)
plt.ylabel("Error between compressed and original image")
plt.xlabel("Number of Singular Values")
plt.grid()
plt.show()
````
